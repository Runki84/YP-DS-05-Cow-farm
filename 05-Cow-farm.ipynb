```python
def cat_plot(data, title):
    plt.pie(
        data.value_counts(),
        labels=data.unique(),
        autopct='%1.1f%%')
    plt.title(title)
    plt.grid()
    plt.show()
```

## 5. Обучение моделей линейной регрессии

Поскольку у нас одну и ту же процедуру необходимо потворить три раза, то проще всего для этого написать функцию, которая будет обучать модель и верифицировать данные в зависимости от разных наборов тренировочных.
Но для начала необходимо будет создать пустой датафрейм, в который будем записывать результаты всех моделей, чтобы можно было извлекать из него данные для сравнения.

```python
# создадим датафрейм для записи результатов работы моделей
results = pd.DataFrame(index=range(3), columns=['predict', 'r2', 'res_mean', 'mse', 'mae', 'rmse' ])
```

```python
# запишем в качестве глобальных отдельно каждую из трех моделей линейной регрессии,
# чтобы можно было с ними работать поотдельности за рамками функции
model_lr_1 = LinearRegression()
model_lr_2 = LinearRegression()
model_lr_3 = LinearRegression()

# запишем функцию, которая по шагам будет готовить данные и обучать модели
# параметр data - для датафрейма с которым будем работать, параметр num для номера модели линейной регрессии
def lin_reg (data, num):

    # Шаг 1. Разбивка данных на тенировочные и валидационные
    RANDOM_STATE = 42
    X = data.drop('yeld_kg_y', axis=1)
    y = data['yeld_kg_y']

    X_train, X_valid, y_train, y_valid = train_test_split(
        X,
        y,
        random_state=RANDOM_STATE
    )
    print(f'Размер выборок модели № {num}:')
    print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
    print()

    # Шаг 2. Кодировка категориальных признаков
    # вместо перечисления названия столбцов, соберем универсальный метод по типу данных в столбце
    # это позволит переиспользовать код в дальнейших работах
    cat_col_names = X_train.select_dtypes('object').columns
    encoder = OneHotEncoder(drop='first', sparse=False)
    X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
    X_valid_ohe = encoder.transform(X_valid[cat_col_names])
    encoder_col_names = encoder.get_feature_names()

    # Шаг 3. Масштабирование (стандартизация) количественных данных
    # аналогично соберем универсальный метод для количественных данных
    num_col_names = X_train.select_dtypes('number').columns
    scaler = StandardScaler()
    X_train_scaled =  scaler.fit_transform(X_train[num_col_names])
    X_valid_scaled = scaler.transform(X_valid[num_col_names])

    # Шаг 4. Сборка выборок
    X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
    X_valid_ohe = pd.DataFrame(X_valid_ohe, columns=encoder_col_names)

    X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
    X_valid_scaled = pd.DataFrame(X_valid_scaled, columns=num_col_names)

    X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
    X_valid = pd.concat([X_valid_ohe, X_valid_scaled], axis=1)
    print(f'Проверка данных для обучения модели № {num}:')
    print(X_train.head())
    print()

    # Шаг 5. Оценка корректности количественных данных
    print(f'Оценка корректности количественных данных для модели № {num}')
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
    X_train[num_col_names].plot(kind='hist', bins=20, ax=axes[0])
    axes[0].set_title('Гистограмма распределения')
    X_train[num_col_names].plot(kind='box', ax=axes[1], rot=45)
    axes[1].set_title('Разброс значений признаков')
    plt.show()
    print()

    # Шаг 6. Обучение моделей линейной регрессии
    num -= 1 # здесь нужно уменьшить значение num на 1, так как индексы начинаются с 0

    # дальше запускаем ветвеление по условиям и раздельное обучение каждой модели
    # это понадобится чтобы выбрать нужную модель для финальной работы
    if num == 0:
        global model_lr_1
        model_lr_1 = LinearRegression()
        model_lr_1.fit(X_train, y_train)
        results.at[num, 'predict'] = model_lr_1.predict(X_valid)
    elif num == 1:
        global model_lr_2
        model_lr_2 = LinearRegression()
        model_lr_2.fit(X_train, y_train)
        results.at[num, 'predict'] = model_lr_2.predict(X_valid)
    elif num == 2:
        global model_lr_3
        model_lr_3 = LinearRegression()
        model_lr_3.fit(X_train, y_train)
        results.at[num, 'predict'] = model_lr_3.predict(X_valid)

    # Шаг 7. Оценка качества модели
    # посчитаем значения R2, MSE, MAE и RMSE и запишем в таблицу с рузальтатами,
    # чтобы было удобнее их значнеия сравнить для разных моделей
    results.at[num, 'r2'] = r2_score(y_valid, results['predict'].iloc[num])
    results.at[num, 'mse'] = mean_squared_error(y_valid, results['predict'].iloc[num])
    results.at[num, 'rmse'] = mean_squared_error(y_valid, results['predict'].iloc[num], squared=False)
    results.at[num, 'mae'] = mean_absolute_error(y_valid, results['predict'].iloc[num])
    # тут придется каждую метрику записать в переменную для удобства вывода на экран
    r2 = results['r2'].iloc[num]
    mse = results['mse'].iloc[num]
    rmse = results['rmse'].iloc[num]
    mae = results['mae'].iloc[num]
    # выведем на экран значения метрик
    print(f'R2 модели № {num + 1}: {r2}')
    print(f'MSE модели № {num + 1}: {mse}')
    print(f'RMSE модели № {num + 1}: {rmse}')
    print(f'MAE модели № {num + 1}: {mae}')
    print()

    # Посчитаем остатки
    residuals = y_valid - results['predict'].iloc[num]
    results.at[num, 'res_mean'] = residuals.mean()
    # для удобства анализа выведем значения остатков в виде гистрограммы и графика рассеяния
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
    axes[0].hist(residuals)
    axes[0].set_title('Гистограмма распределения остатков')
    axes[0].set_xlabel('Остатки')
    axes[1].scatter(results['predict'].iloc[num], residuals)
    axes[1].set_xlabel('Предсказания модели')
    axes[1].set_ylabel('Остатки')
    axes[1].set_title('Анализ дисперсии')
    plt.show()
```

## 6. Обучение модели логистической регрессии

Для обучения модели логистической регрессии, нам необходимо решить задачу классификации: определить на основании входных признаков, к какой категории потенциально может относиться молоко покупаемых коров, к "вкусному" или "не вкусному".
При этом у нас есть прямая задача от заказчика, отобрать только коров которые точно будут давать вкусное молоко. Иными словами мы можем допустить ложно-отрицательные ошибки, но не можем допустить ложно-положительные.

Для обучения модели, нам необходимо отобрать только те признаки которые могут оказать влияние на вкус молока, и имеющие корреляционную зависимость с целевым признаком - вкусом молока.
Основные параметры, влияющие на вкус молока, это его питательные свойства, то есть жирность и процентное содержание белка. Влияние этих признаков научно достоверный факт. Также, научно подтвержденным фактом, влияющим на вкус молока, является состав кормов. А состав кормов напрямую зависит от состава растительности пастбища, то есть тип пастбища тоже коррелирует со вкусовыми качествами молока.
Поэтому из выборки мы однозначно убираем следующие признаки:
 - id коровы - этот признак в принципе не имеет никакой корреляции со вкусом молока
 - ЭКЕ. Этот признак влияет на продуктивность, но не влияет на вкус молока, так как он описывает энергетическую ценность кормов, но не их питательный состав.
 - породу коров и породу отца-быка мы исключаем, так как матрица корреляции показывает 0-е значение. Аналогично можем поступить и с именем отца быка, так как этот параметр не влияет на вкусовые качества молока
 - не смотря на наличие положительной корреляции, мы уберем из выборки и производительность молока. Опять же научно достоверным является факт отрицательной корреляции между объемом удоя и вкусовыми качествами молока: чем выше удой, тем хуже вкусовые качества молока. Но в нашей выборке это правило нарушено и мы видим положительную корреляцию из-за сильного влияния кормов как на удой, так и на вкусовые качества. Поэтому, чтобы избежать ошибок прогнозирования, удой из обучения необходимо убирать.

```python
# создадим рабочий датафрейм для манипуляций и обучения модеди на основании основного - data_farm
# намеренно не объяединяю с data_dad, так как эти признаки исключаем
farm_log = data_farm
# cнова проведем категорирование СПО и возведение в квадрат ЭКЕ
farm_log['spr'] = farm_log['spr'].apply(lambda x: 'менее 0.92' if x < 0.92 else 'более 0.92')
farm_log['efu2'] = farm_log['efu'] ** 2
# переведем в числовой формат целевой признак: 1 - успешный результат (вкусно), 0 - не успешный результат (не вкусно)
farm_log['milk_taste'] = farm_log['milk_taste'].apply(lambda x: 1 if x == 'вкусно' else 0)
# уберем лишние признаки
farm_log = farm_log.drop(['id', 'yeld_kg_y', 'efu', 'breed', 'dad_bull_breed'], axis=1)
```

```python
RANDOM_STATE = 42
X = farm_log.drop('milk_taste', axis=1)
y = farm_log['milk_taste']

X_train, X_valid, y_train, y_valid = train_test_split(
    X,
    y,
    random_state=RANDOM_STATE
)
print(f'Размер выборок:')
print(X_train.shape, X_valid.shape, y_train.shape, y_valid.shape)
print()

# Шаг 2. Кодировка категориальных признаков
cat_col_names = X_train.select_dtypes('object').columns
encoder = OneHotEncoder(drop='first', sparse=False)
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
X_valid_ohe = encoder.transform(X_valid[cat_col_names])
encoder_col_names = encoder.get_feature_names()

# Шаг 3. Масштабирование (стандартизация) количественных данных
num_col_names = X_train.select_dtypes('number').columns
scaler = StandardScaler()
X_train_scaled =  scaler.fit_transform(X_train[num_col_names])
X_valid_scaled = scaler.transform(X_valid[num_col_names])

# Шаг 4. Сборка выборок
X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_valid_ohe = pd.DataFrame(X_valid_ohe, columns=encoder_col_names)

X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_valid_scaled = pd.DataFrame(X_valid_scaled, columns=num_col_names)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_valid = pd.concat([X_valid_ohe, X_valid_scaled], axis=1)

# Шаг 6. Обучение моделей линейной регрессии
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_valid)

# Шаг 7. Оценка качества модели
accuracy = accuracy_score(y_valid, y_pred)
recall = recall_score(y_valid, y_pred)
precision = precision_score(y_valid, y_pred)

print(f'Значение Accuracy: {round(accuracy, 3)}')
print(f'Значение Recall: {round(recall, 3)}')
print(f'Значение Precision: {round(precision, 3)}')

# Построение матрицы ошибок
cm = confusion_matrix(y_valid, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
plt.ylabel('Истинные значения')
plt.xlabel('Предсказанные значения');
```

На валидационной выборке в 157 значений модель предсказала:

 - Истинно-отрицательные значения (предсказано не вкусное молоко) - 24 раз
 - Ложно-отрицательные значения (предсказано не вкусное молоко) - 12 раз
 - Ложно-положительные значения (предсказано вкусное молоко) - 45 раз
 - Истинно-положительные значения (предсказано вкусное молоко) - 76 раз

В целом предсказания истинно-положительных значений довольно высокие

По условиям нам нужно, чтобы молоко **всегда** было вкусным, значит нам нельзя допускать ложно-положительных значений (чтобы модель ошибочно предсказывала "вкусное" молоко), но допустимы ложно-отрицательные значения (когда модель вкусное молоко определяет как не вкусное), а значит нам надо избегать ошибок 1 рода, но допустимы ошибки 2 рода. То есть нам необходимо максимизировать определение 0-вых значений (отрицательных). Соответственно задача свести к максимуму значение метрики Precision и свести к минимуму значение метрики Recall

```python
# построим кривую соотношения Recall и Precision, для определенич оптимальных параметров
proba_valid = clf.predict_proba(X_valid)
precision, recall, thresholds = precision_recall_curve(y_valid, proba_valid[:, 1])

plt.figure(figsize=(6, 6))
plt.step(recall, precision, where='post')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.xticks(np.arange(0, 1, 0.1))
plt.yticks(np.arange(0, 1, 0.1))
plt.grid()
plt.title('Кривая Precision-Recall')
plt.show()
```
